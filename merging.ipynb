```python
df1 = pd.DataFrame(
    {
        "time": [
            pd.Timestamp("2016-05-25 13:31:00.023"),
            pd.Timestamp("2016-05-25 13:32:00.023"),
            pd.Timestamp("2016-05-25 13:33:00.030"),
            pd.Timestamp("2016-05-25 13:42:00.041"),
            pd.Timestamp("2016-05-25 13:35:00.048"),
            pd.Timestamp("2016-05-25 13:38:00.049"),
            pd.Timestamp("2016-05-25 13:40:00.072"),
            pd.Timestamp("2016-05-25 13:50:00.075")
        ],
    
        "time2": [
            pd.Timestamp("2016-05-25 14:31:00.023"),
            pd.Timestamp("2016-05-25 14:32:00.023"),
            pd.Timestamp("2016-05-25 14:33:00.030"),
            pd.Timestamp("2016-05-25 14:42:00.041"),
            pd.Timestamp("2016-05-25 14:35:00.048"),
            pd.Timestamp("2016-05-25 14:38:00.049"),
            pd.Timestamp("2016-05-25 14:40:00.072"),
            pd.Timestamp("2016-05-25 14:50:00.075")
        ]})
df1
```

```python
df1['MIN_YRMO'] = df1[['time', 'time2']].min(axis=1, skipna=False)
df1['MAX_YRMO'] = df1[['time', 'time2']].max(axis=1, skipna=False)

    
df1
```
